```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

excel_path = "./hate_crime_quarter.xlsx"
df = pd.read_excel (excel_path)

df
```

```python
df=df[["Precinct","Race"]]
df
```

```python
excel_path = "./racial_crime_for_race.xlsx"
racial_crime = pd.read_excel (excel_path)

racial_crime.head()
```

```python
excel_path = "./precincts_zips.xlsx"
neighborhoods = pd.read_excel (excel_path)
df = df.rename(columns = {"Precinct":"precinct_number"})
neighborhoods.head()
```

```python
zipcode_crime = pd.merge(df,neighborhoods, on = "precinct_number" , how = "left" )
zipcode_crime.head()
```

```python
zipcode_crime = zipcode_crime[["Race","zipcode"]]
zipcode_crime.head()
```

```python
black = zipcode_crime[zipcode_crime["Race"] == "Black"]
white = zipcode_crime[zipcode_crime["Race"] == "White"]
asian = zipcode_crime[zipcode_crime["Race"] == "Asian"]
hispanic = zipcode_crime[zipcode_crime["Race"] == "Hispanic"]

asian
```

```python
black=black.groupby('zipcode').count()
black=black.reset_index()
white=white.groupby('zipcode').count()
white=white.reset_index()
asian=asian.groupby('zipcode').count()
asian=asian.reset_index()
hispanic=hispanic.groupby('zipcode').count()
hispanic=hispanic.reset_index()
hispanic
```

```python
merge1 = pd.merge(
    black, white, on="zipcode", how="outer", suffixes=("_Black_Offender", "_White_Offender"))
merge1
```

```python
merge2 = pd.merge(asian,hispanic,on = "zipcode", how = "outer",suffixes=("_Asian_Offender", "_Hispanic_Offender"))
zipcode_hate = pd.merge(merge1,merge2, on="zipcode", how = "outer")
zipcode_hate=zipcode_hate.fillna(0)
zipcode_hate
```

```python
new_df = pd.merge(zipcode_hate,racial_crime, on="zipcode", how = "right")
new_df=new_df.fillna(0)
new_df.head()
```

```python
import matplotlib.colors
zipcodes = new_df["zipcode"]
plt.figure(figsize=(20,7))
x_axis = np.arange(len(zipcodes))
y  = new_df["percentage_black"]

tick_locations = [value+0.4 for value in x_axis]

c = new_df["Race_Black_Offender"]
df = pd.DataFrame({"x":x_axis,"y":y,"c":c})

cmap = plt.cm.rainbow
norm = matplotlib.colors.Normalize(vmin=0, vmax=15)

fig, ax = plt.subplots()
ax.bar(df.x, df.y, color=cmap(norm(df.c.values)))
ax.set_xticks(df.x)

sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])  # only needed for matplotlib < 3.1
fig.colorbar(sm)


plt.xticks(tick_locations, new_df["zipcode"], rotation="vertical")
plt.ylabel('Percentage of population - black')
plt.title('Percentage of Total Population - black and number of hate crime offenders - black')
plt.text(50, 35, 'Number of Hate Crimes', fontsize=10)
plt.savefig("./Black_Crime.png")
plt.show()
```
