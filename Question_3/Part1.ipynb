```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

excel_path = "./hate_crime_quarter.xlsx"
df = pd.read_excel (excel_path)

df
```

```python
df['Bias-Motivation'].unique()
df['Bias-Motivation'].value_counts()
```

```python
df['Bias-Motivation'] = df['Bias-Motivation'].replace(
    {'BLACK': 'Anti-Black', 'Anti-asian': 'Anti-Asian','Anti-female  homosexual (Lesbian)':'Anti-Female Homosexual (Lesbian)','Anti-Female homosexual (Lesbian)':'Anti-Female Homosexual (Lesbian)'})
df['Bias-Motivation'].value_counts()
```

```python
df['Bias-Motivation'] = df['Bias-Motivation'].replace(
    {'Anti-other religion': 'Anti-Other Religion', 'Anti-Other Anti-Other Religion': 'Anti-Other Religion','Anti-male homosexual (Gay)':'Anti-Male Homosexual (Gay)','Anti-Male homosexual (Gay)':'Anti-Male Homosexual (Gay)'})
df['Bias-Motivation'].value_counts()
```

```python
df['Bias-Motivation'] = df['Bias-Motivation'].replace(
    {'Anti-Islamic': 'Anti-Islamic (Muslim)', 'Anti-other Ethnicity': 'Anti-Other Ethnicity','Anti-Islam (Muslim)':'Anti-Islamic (Muslim)'})
#df['Bias-Motivation'].value_counts()

df['Bias-Motivation'].value_counts()
```

```python
race_bias = ['Anti-Black','Anti-White','Anti-Other Ethnicity','Anti-Asian','Anti-Hispanic','Anti-Arab']
religion_bias = ['Anti-Jewish','Anti-Islamic (Muslim)','Anti-Other Religion','Anti-Catholic','Anti-Mormon','Anti-Multi Religious Groups']
sexual_bias = ['Anti-Male Homosexual (Gay)','Anti-LGBT','Anti-Female Homosexual (Lesbian)']
gender_id_bias = ['Anti-Transgender','Anti-Gender Non-Conforming']
disability_bias = ['Anti-physical disability']
```

```python
#df["Bias-Motivation"]
excel_path = "./crime_clean.xlsx"
new_df = pd.read_excel (excel_path)

new_df


```

```python
bias_group = []
for bias in new_df['Bias-Motivation']:
    if bias in race_bias:
        bias_group.append("Racial/Ethnicity/Ancestry Bias")
    elif bias in religion_bias:
        bias_group.append("Religious Bias")
    elif bias in sexual_bias:
        bias_group.append("Sexual Orientation Bias")
    elif bias in gender_id_bias:
        bias_group.append("Gender Identity Bias")
    elif bias in disability_bias:
        bias_group.append("Disability Bias")
    else:
        bias_group.append("Other Bias")
bias_group
```

```python

new_df["Bias-Group"]=bias_group
new_df.to_excel('./crime_bias_group.xlsx')
new_df
```

```python
bias_df=new_df.groupby('Bias-Group')["Bias-Group"].count()

bias_pie = bias_df.plot(kind="pie",autopct="%1.1f%%",title = "% of Hate Crimes in NYC by Bias type", shadow=True, startangle=140)
plt.savefig("./Bias-Group.png")
plt.show()
```

```python
excel_path = "./Summary_Bias_FBI.xlsx"
federal_df = pd.read_excel (excel_path)

federal_df
```

```python
federal_bias_df=federal_df.groupby('Bias Type')["Incidents"].sum()

federal_bias_pie = federal_bias_df.plot(kind="pie",autopct="%1.1f%%",title = "% of Hate Crimes in the US by Bias type", shadow=True, startangle=20)
plt.savefig("./Federal_Bias-Group.png")
plt.show()
```

```python
excel_path = "./State_FBI_Summary.xlsx"
state_df = pd.read_excel (excel_path)
state_df["Total"] = state_df["Racial/Ethnicity/Ancestry Bias"]+state_df["Religion"]+state_df["Sexual Orientation"]+state_df["Disability"]+state_df["Gender"]+state_df["Gender Identity"]
state_df_percentage = state_df
state_df_percentage["Racial/Ethnicity/Ancestry Bias"]=state_df_percentage["Racial/Ethnicity/Ancestry Bias"]/state_df_percentage["Total"]
state_df_percentage["Religion"]=state_df_percentage["Religion"]/state_df_percentage["Total"]
state_df_percentage["Sexual Orientation"]=state_df_percentage["Sexual Orientation"]/state_df_percentage["Total"]
state_df_percentage["Disability"]=state_df_percentage["Disability"]/state_df_percentage["Total"]
state_df_percentage["Gender"]=state_df_percentage["Gender"]/state_df_percentage["Total"]
state_df_percentage["Gender Identity"]=state_df_percentage["Gender Identity"]/state_df_percentage["Total"]
state_df_percentage
```

```python
plt.figure(figsize=(20,3))

x_axis = np.arange(len(state_df))
tick_locations = [value+0.4 for value in x_axis]



p1 = plt.bar(x_axis, state_df_percentage['Racial/Ethnicity/Ancestry Bias'])
p2 = plt.bar(x_axis, state_df_percentage["Religion"],
             bottom=state_df_percentage['Racial/Ethnicity/Ancestry Bias'])
p3 = plt.bar(x_axis, state_df_percentage["Sexual Orientation"],
             bottom=(state_df_percentage["Racial/Ethnicity/Ancestry Bias"]+state_df["Religion"]))
p4 = plt.bar(x_axis, state_df_percentage["Gender Identity"],
             bottom=(state_df_percentage["Racial/Ethnicity/Ancestry Bias"]+state_df["Religion"]+state_df["Sexual Orientation"]))

plt.xticks(tick_locations, state_df["State"], rotation="vertical")
plt.ylabel('Percentage of Total Incidents by State, 4 Major Bias Groups')
plt.title('Incidents by State')
plt.legend((p1[0], p2[0],p3[0], p4[0]), ('Racial/Ethnicity/Ancestry Bias', "Religious Bias",
                                         "Sexual Orientation Bias", "Gender Identity Bias"))
plt.savefig("./State_Bias-Group_Percentage.png")
plt.show()
```
